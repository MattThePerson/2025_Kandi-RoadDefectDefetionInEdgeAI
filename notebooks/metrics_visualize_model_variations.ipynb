```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
metrics_df = pd.read_csv("analytics/metrics_table.csv")
metrics_df = metrics_df[["model_name", "run_group", "quantized", "imgsz", "mAP50_mean"]]
metrics_df.groupby("model_name")
metrics_df
```

```python
# 
d = {}
for k, v in dict(tuple(metrics_df.groupby("model_name"))).items():
    v["value"] = v["mAP50_mean"]
    v["col"] = v["quantized"] + "_" + v["imgsz"].astype(str)
    v = v[["col", "value"]]
    var2value = dict(v.to_numpy())
    # print(k)
    # print(var2value)
    d[k] = var2value
    # break
df = pd.DataFrame(d).T
df = df[["original_640", "original_320", "quant_640", "quant_320"]]
df.to_csv("analytics/model_variations.csv")
df
```
